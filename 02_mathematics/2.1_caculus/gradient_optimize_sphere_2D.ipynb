```python
# sphere function
def func(x, y):
    return x**2 + y**2
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LogNorm

xmin, xmax, xstep = -4.5, 4.5, .2
ymin, ymax, ystep = -4.5, 4.5, .2

x, y = np.meshgrid(np.arange(xmin, xmax + xstep, xstep), np.arange(ymin, ymax + ystep, ystep))
z = func(x, y)

# 3D Surface Plot
fig = plt.figure(figsize=(8, 5))
ax = plt.axes(projection='3d', elev=0, azim=-40)

ax.plot_surface(x, y, z, norm=LogNorm(), rstride=1, cstride=1, 
                edgecolor='none', alpha=.9, cmap=plt.cm.jet)

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')

ax.set_xlim((xmin, xmax))
ax.set_ylim((ymin, ymax))

plt.show()
```

```python
import numpy as np

# sphere function
def func(x, y):
    return x**2 + y**2

# o hm trung tm
def gradient(f, x, y, e=1.0e-4):
    dx = (f(x+e/2, y) - f(x-e/2, y)) / e
    dy = (f(x, y+e/2) - f(x, y-e/2)) / e
    return dx, dy
```

```python
import random

# 1. set (x,y) randomly
x = random.randint(-100, 100)
y = random.randint(-100, 100)
print(x, y)

# params
num_iterations = 80
step = 1

# optimize
for _ in range(num_iterations):
    # 2. compute the derivative at (x,y)
    dx, dy = gradient(func, x, y)
    
    # get sign of dx
    sign_x = np.sign(dx)
    sign_y = np.sign(dy)
        
    # update
    x = x - sign_x*step
    y = y - sign_y*step
    
    print(x, y, func(x,y))
```
