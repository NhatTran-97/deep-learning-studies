```python
'''
f(x) = x^2 + 3x
g(u) = u^2 + x + 2

compute the derivative of g(f(x))
''' 

def f_func(x):
    return x**2 + 3*x

def g_func(x):
    return x**3 + x + 2
```

```python
# plot g_func

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-50, 50, 0.1)
g = g_func(x)

plt.plot(x, g)
plt.show()
```

```python
def f_func(x):
    return x**2 + 3*x

def g_func(x):
    return x**3 + x + 2

# đạo hàm trung tâm
def gradient(func, x, e=1.0e-4):     
    return (func(x + e/2) - func(x - e/2)) / e
```

```python
# minimize the composite function g(f(x))

import random
import numpy as np

num_steps = 100
step = 0.02

# 1. set x randomly
x = -2.61 #random.uniform(-4.4, 1.5)
print(x)

for _ in range(num_steps):
    # 2. compute derivative
    fx = f_func(x)    
    df_dx = gradient(f_func, x)
    dg_df = gradient(g_func, fx)
    dg_dx = dg_df*df_dx
    
    # 3. get sign
    sign = np.sign(dg_dx)
    
    # 4 update
    x = x - sign*step
    print(x, g_func(f_func(x)))
```
