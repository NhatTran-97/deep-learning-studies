```python
def bubbleSort(arr):
    n = len(arr)
    step = 1
    for i in range(n-1):
        for j in range(0, n-i-1):
            print("step: ", step)
            step = step + 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
```

```python
array = [4,6, 7, 9]
bubbleSort(array)
print(array)
```

```python
def bubbleSort_optimize(arr):
    n = len(arr)
    step = 1
    for i in range(n-1):
        swapped = False
        for j in range(0, n-i-1):
            print("step: ", step)
            step = step + 1
            if arr[j] > arr[j + 1]:
                swapped = True
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
        if not swapped:
            return
```

```python
array = [4,6, 7, 9]
bubbleSort_optimize(array)
print(array)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create the data
x = np.linspace(1, 100, 100)  # Represents the size of the input data (n)
y_n = x  # O(n) linear time
y_log_n = np.log2(x)  # O(log n) logarithmic time
y_n_log_n = x * np.log2(x)  # O(n log n) quasilinear time
y_n_square = x**2  # O(n^2) quadratic time

# Create the plot
plt.plot(x, y_n, label='O(n)')
plt.plot(x, y_log_n, label='O(log n)')
plt.plot(x, y_n_log_n, label='O(n log n)')
plt.plot(x, y_n_square, label='O(n^2)')

# Set the title and axis labels
plt.title('Time Complexity')
plt.xlabel('Input size (n)')
plt.ylabel('Time (steps)')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
def binarySearchRecursive(list_data, low, high, search_value):
    if high >= low:

        mid = low + (high - low) // 2

        if list_data[mid] == search_value:
            return mid

        elif list_data[mid] > search_value:
            return binarySearchRecursive(list_data, low, mid-1, search_value)

        else:
            return binarySearchRecursive(list_data, mid + 1, high, search_value)

    else:
        return -1
```

```python
numbers = [2, 3, 4, 10, 40]
search_value = 10
result = binarySearchRecursive(numbers, 0, len(numbers)-1, search_value)
if result != -1:
  print("Element is present at index", result)
else:
  print("Element is not present in array")
```
