```python
name = "Duy Nhat"
print(name)
```

```python
# Iterate a string
for character in name:
    print(character)
```

```python
# Iterate a string
for index in range(len(name)):
    print(name[index])
```

```python
name_of_lab = 'EIU FabLab (fablab@eiu.edu.vn) - Eastern International University.'
atpos = name_of_lab.find('@')
print("atpos: ", atpos)
sppos = name_of_lab.find(')',atpos)
print("sppos: ", sppos)
host = name_of_lab[atpos+1 : sppos]
print("host: ", host)
```
